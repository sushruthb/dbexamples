```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("DeltaLakeLocal") \
    .config("spark.sql.extensions", "io.delta.sql.DeltaSparkSessionExtension") \
    .config("spark.sql.catalog.spark_catalog", "org.apache.spark.sql.delta.catalog.DeltaCatalog") \
    .getOrCreate()
```

```python
data = [(1, "IN", 100), (2, "US", 200)]
df = spark.createDataFrame(data, ["id", "region", "amount"])

df.write.format("delta").mode("overwrite").save("/Volumes/workspace/dbricks/db/sales")
```

```python
spark.read.format("delta").load("/Volumes/workspace/dbricks/db/sales").show()
```
